```python
import matplotlib.pyplot as plt

# Normalizar la variable precios
subconjunto_cinturon_sur['Normal Price'] = (subconjunto_cinturon_sur['price'] - subconjunto_cinturon_sur['price'].min()) / (subconjunto_cinturon_sur['price'].max() - subconjunto_cinturon_sur['price'].min())

# Crear los subplots
fig, axis = plt.subplots(2, 2, figsize=(12, 7))

axis[0, 0].hist(subconjunto_cinturon_sur.loc[subconjunto_cinturon_sur['level5'] == 'Getafe']['Normal Price'])
axis[0, 0].set_title('Getafe')

axis[0, 1].hist(subconjunto_cinturon_sur.loc[subconjunto_cinturon_sur['level5'] == 'Alcorcón']['Normal Price'])
axis[0, 1].set_title('Alcorcón')

axis[1, 0].hist(subconjunto_cinturon_sur.loc[subconjunto_cinturon_sur['level5'] == 'Fuenlabrada']['Normal Price'])
axis[1, 0].set_title('Fuenlabrada')

axis[1, 1].hist(subconjunto_cinturon_sur.loc[subconjunto_cinturon_sur['level5'] == 'Leganés']['Normal Price'])
axis[1, 1].set_title('Leganés')

plt.tight_layout()

plt.show()
```
