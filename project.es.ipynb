#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (price/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio y escribe una conclusion acerca de ellos.

Pista: Crea una nueva columna llamada `pps` (price per square) y luego analiza los valore.s

```python
# TODO
import pandas as pd

# Cargar el dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Filtrar las filas para Valdemorillo y Galapagar
valdemorillo_df = ds[ds['level5'] == "Valdemorillo"]
galapagar_df = ds[ds['level5'] == "Galapagar"]

# Calcular el precio por metro cuadrado (price per square, pps)
valdemorillo_df['pps'] = valdemorillo_df['price'] / valdemorillo_df['surface']
galapagar_df['pps'] = galapagar_df['price'] / galapagar_df['surface']

# Calcular el promedio de pps para Valdemorillo y Galapagar
promedio_pps_valdemorillo = valdemorillo_df['pps'].mean()
promedio_pps_galapagar = galapagar_df['pps'].mean()

# Imprimir ambos promedios
print(f"Promedio de precio por metro cuadrado en Valdemorillo: {promedio_pps_valdemorillo:.2f} USD/m2")
print(f"Promedio de precio por metro cuadrado en Galapagar: {promedio_pps_galapagar:.2f} USD/m2")
```

El precio por metro cuadrado de Galapar es de 1606.32 USD mientras que el precio por metro cuadrado es de 1317.95 en Valdemorillo. Por lo tanto podemos decir que el precio por metro cuadrado es más alto en Galapar. Por otra parte, antes hemos visto que el precio medio de ambas poblaciones eran practimanete el mismo, por lo tanto, Galapar debe tener casas de tamaño más pequeño, incrementando así su precio por metro cuadrado.

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Define the populations of interest
poblaciones_cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Create a subset DataFrame with these populations
subset_cinturon_sur = ds[ds['level5'].isin(poblaciones_cinturon_sur)]

# Normalize prices for each population
normalized_prices = {}
for population in poblaciones_cinturon_sur:
    prices = subset_cinturon_sur[subset_cinturon_sur['level5'] == population]['price']
    normalized_prices[population] = (prices - prices.min()) / (prices.max() - prices.min())

# Plot histograms for normalized prices
plt.figure(figsize=(10, 6))
for population, prices in normalized_prices.items():
    plt.hist(prices, bins=20, alpha=0.5, label=population)

plt.title('Normalized Prices Histogram')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Define the populations of interest
poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Create a subset DataFrame with these populations
subset = ds[ds['level5'].isin(poblaciones)]

# Create a grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Iterate over the populations and create scatter plots on subplots
for i, poblacion in enumerate(poblaciones):
    row = i // 2
    col = i % 2
    subset_poblacion = subset[subset['level5'] == poblacion]
    axs[row, col].scatter(subset_poblacion['surface'], subset_poblacion['price'], label=poblacion)
    axs[row, col].set_title(poblacion)
    axs[row, col].set_xlabel('Metros')
    axs[row, col].set_ylabel('Precio')
    axs[row, col].legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Define the populations of interest
poblaciones_cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Create a subset DataFrame with these populations
subset_cinturon_sur = ds[ds['level5'].isin(poblaciones_cinturon_sur)]

# Create a dictionary to store coordinates by population
coordinates_by_population = {}

# Iterate over the populations and add the corresponding coordinates to the dictionary
for poblacion in poblaciones_cinturon_sur:
    subset_population = subset_cinturon_sur[subset_cinturon_sur['level5'] == poblacion]
    coordinates_by_population[poblacion] = {
        'latitude': subset_population['latitude'].tolist(),
        'longitude': subset_population['longitude'].tolist()
    }

# Plot the coordinates of each population on a map
plt.figure(figsize=(10, 8))
for population, coordinates in coordinates_by_population.items():
    plt.scatter(coordinates['latitude'], coordinates['longitude'], label=population)

plt.title('Coordinates of the Cinturón Sur of Madrid')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.legend()
plt.grid(True)
plt.show()
```
